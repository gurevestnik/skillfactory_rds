# 31. Названия фильмов какой студии в среднем самые длинные по количеству символов?
Варианты ответа:
1. Universal Pictures (Universal)
2. Warner Bros
3. Jim Henson Company, The
4. Paramount Pictures
5. Four By Two Productions

```python
data = pd.read_csv('cinema.csv')
# выборка по подсчету символов
data['len'] = data.original_title.apply(lambda pp: len(pp))


import collections
c = collections.Counter()

for gg in data.imdb_id:
    dat = data[data.imdb_id == gg]
    prod_c = dat.production_companies.iloc[0]
    d_len = dat.len.iloc[0]
    for aa in prod_c.split('|'):
        c[aa] += d_len

for aa, l in c.items():
    dat = data[data.production_companies.str.contains(aa, na = False, regex=False)]
    count = dat.original_title.count()
    c[aa] = l / count

print(c.most_common(1))
```

# 35. Какие актеры чаще всего снимаются в одном фильме вместе
Варианты ответа:
1. Johnny Depp & Helena Bonham Carter
2. Hugh Jackman & Ian McKellen
3. Vin Diesel & Paul Walker
4. Adam Sandler & Kevin James
5. Daniel Radcliffe & Rupert Grint

```python
data = pd.read_csv('cinema.csv')

import collections
from itertools import combinations
c = collections.Counter()

actor = []

for pp in data.cast:
    for gg in pp.split('|'):
        actor.append(gg)
    actor.sort()
    dat = list(combinations(actor, 2))
    actor.clear()
    for i in dat:
        c[i] +=1

print(c.most_common(3))
```
